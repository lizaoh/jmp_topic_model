# Topic Explorer Notebook Tutorial

The InPhO Topic Explorer features a powerful interactive coding environment that enables direct manipulation of the corpus and models, in contrast to the web visualization.

When you run the `vsm notebook` command, several things happen:
1.  The Topic Explorer creates a new folder called `notebooks` and places several files inside it:
     -  **corpus.py** contains Python code which imports the Corpus object and a function that will load models trained on your corpus. It gathers this information from your `CORPUS.ini` file, where `CORPUS` is the name of the folder you prepared. More information on this special file is below.
     -  **Topic Explorer Tutorial.ipynb** is this file, which provides the skeleton documentation for interacting with the models.
     -  Other **ipynb** files, containing analyses that can be re-run on your own corpora.
2.  The Topic Explorer launches a [Jupyter Notebook](http://jupyter.org/). This allows you to program in Python on your local computer using the browser, rather than a terminal or other program.

The `normed` argument tells the program whether to weight each document by its length (`normed_sum=True`) or to consider them all equally (`normed_sum=False`).

### Comparing documents with `v.dist()`

Topic models give us a way to compare the siimilarity between two documents. To do this, we use `v.dist()`:

```python
v.dist(v.labels[0], v.labels[1])
```

#### Alternative distance measures
By default, the Topic Explorer uses the Jensen-Shannon Distance to calculate the distance between documents. The Jensen-Shannon Distance (JSD) is a symmetric measure based on information theory that characterizes the difference between two probability distributions.

However, several alternate methods are built into the `vsm.spatial` module. These include the Kullback-Leibler Divergence, which is an asymmetric component of the JSD and is used in [Murdock et al. (2017)](http://arxiv.org/abs/1509.07175) to characterize the cognitive surprise of a new text, given previous texts.

Rather than using the JSD and assuming symmetric divergence between items, we assume that the second document is encountered after the first, effectively measuring text-to-text divergence.

```python
help(v.dist)
```

```python
# first import KL divergence:
from vsm.spatial import KL_div

# calculate KL divergence from the first document to the second
print("First to second", v.dist(v.labels[0], v.labels[1], dist_fn=KL_div))

# calculate KL divergence from the second document to the first, highlighting asymmetry:
print("Second to first", v.dist(v.labels[1], v.labels[0], dist_fn=KL_div))
```

## K-means clusters using JSD values

```python
import numpy as np
import random
import networkx as nx
from pyvis.network import Network
from sklearn.manifold import MDS
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
dois = filtered_df["DOI"].astype(str).str.strip().tolist()
idx = {d:i for i,d in enumerate(dois)}
n = len(dois)

D = np.zeros((n, n), dtype=float)

for _, r in edges_df.iterrows():
    i = idx.get(str(r["doi_source"]).strip())
    j = idx.get(str(r["doi_target"]).strip())
    if i is None or j is None:
        continue
    D[i, j] = float(r["jsd"])

# If edges_df is directional duplicates, symmetrize:
D = 0.5 * (D + D.T)
np.fill_diagonal(D, 0.0)
```

```python
X = MDS(n_components=2, dissimilarity="precomputed", random_state=42).fit_transform(D)  # 2 components for 2D
```

## Calculates similarity based on JSD values

```python
edges_df["jsd_sim"] = np.exp(-edges_df["jsd"])
```

```python
edges_top5 = (
    edges_df
    .groupby("doi_source", group_keys=False)
    .apply(lambda x: x.nsmallest(5, "jsd"))
)

min_jsd_sim = edges_top5["jsd_sim"].min()
max_jsd_sim = edges_top5["jsd_sim"].max()

edges_top5["scaled_jsd_sim"] = (
    (edges_top5["jsd_sim"] - min_jsd_sim) /
    (max_jsd_sim - min_jsd_sim)
)

edges_top5["edge_width"] = 1 + edges_top5["scaled_jsd_sim"] * 8
```

## Setting up network

```python
# Scales number of citations to make node size differences more visible
cites = filtered_df["crossref_ref_by_count"].fillna(0)

log_cites = np.log1p(cites)
scaled = (log_cites - log_cites.min()) / (log_cites.max() - log_cites.min())
filtered_df["node_size"] = 15 + 50 * scaled
```

```python
G = nx.Graph()

# Adds nodes
for file in filtered_df["file_name"]:
    G.add_node(file)
    
# Adds edges
for _, r in edges_top5.iterrows():
    G.add_edge(
        r["source"],
        r["target"],
        value=r["edge_width"],
        title=f"JSD: {r['jsd']:.3f}"
    )

# Computes force layout
pos = nx.spring_layout(G, seed=42, k=0.5)

# Extracts coordinates from force layout
papers = list(G.nodes())
X_force = np.array([pos[p] for p in papers])
```

```python
"""
Computes silhouette score for 2-10 clusters
(this is a measure of how well each paper fits within its
assigned cluster relative to the closest cluster)
"""
scores = []
ks = range(2, 11)

for k in ks:
    labels = KMeans(n_clusters=k, random_state=42, n_init=20).fit_predict(X_force)
    score = silhouette_score(X_force, labels)
    scores.append(score)
    print(f"k={k}, silhouette={score:.3f}")
```

```python
# Saves clusters for 3, 5, and 10 clusters in the network
k_values = [3, 5, 10]

cluster_results = {}

for k in k_values:
    labels = KMeans(n_clusters=k, random_state=42).fit_predict(X_force)
    cluster_results[k] = labels
```

## Makes network based on JSD values

```python
# Runs k-means on the force layout
k = 10

# Assigns each paper to a cluster
labels = cluster_results[k]
cluster_map = dict(zip(papers, labels))
filtered_df["cluster"] = filtered_df["file_name"].map(cluster_map)

cluster_colors = [
    "#1f77b4",  # strong blue
    "#ff7f0e",  # bright orange
    "#2ca02c",  # green
    "#d62728",  # red
    "#9467bd",  # purple
    "#8c564b",  # brown
    "#e377c2",  # magenta
    "#7f7f7f",  # gray
    "#bcbd22",  # olive
    "#17becf"   # cyan
]
clusters = sorted(set(labels))
color_map = {
    c: cluster_colors[i]
    for i, c in enumerate(sorted(clusters))
}

# Adds visual attributes
metadata = (
    filtered_df
    .set_index("file_name")
    .to_dict(orient="index")
)

for paper in papers:
    x, y = pos[paper]
    info = metadata[paper]

    G.nodes[paper]["x"] = x * 1000
    G.nodes[paper]["y"] = y * 1000
    G.nodes[paper]["physics"] = False
    G.nodes[paper]["size"] = info["node_size"]
    G.nodes[paper]["color"] = color_map[cluster_map[paper]]
    G.nodes[paper]["title"] = (
        f"{info['title']}\n"
        f"{info['author']}\n{info['issued_year']}\n"
        f"Citations: {info['crossref_ref_by_count']}"
    )

# Visualizes network using PyVis
nt = Network(
    height="600px",
    width="1000px",
    notebook=True,
    cdn_resources="in_line"
)

nt.from_nx(G)
nt.show("force_kmeans_network.html")
```

# Using Python's Help System

There are many other functions in the InPhO Topic Explorer and the associated `vsm` library. These are extensively documented within the code. 

One little-known feature about Python is its capacity for introspection: by using the `help()` method, one can find out all methods and properties of an object. For example, if one wanted to know what methods could be called on their corpus object, you could run:

```python
help(c)
```

You can also get help on particular methods. For example, there are many arguments to `v.topics()` beyond `print_len`. These can be seen by calling `help(v.topics)` without parentheses after `v.topics`:

```python
help(v.topics)
```

Calling `help(v.topics())` *with* parentheses will return help for the object reutrned by `v.topics()`, which is a `DataTable`:

```python
help(v.topics())
```

It is important to emphasize that this functionality can be used with any python library, including the standard library. For example, one could look at all the functions included in the `math` library by using:

```python
import math
help(math.log)
```

For emphasis, calling `help(math.log(3))` will return the documentation for `float`. Why? First `math.log(3)` will be evaluated, then `help()` will be called on that result: `help(1.0986122886681098)`

```python
help(math.log(3))
```

## `help()` and `?`

Alternatively, you can place a `?` before any function to receive help in a separate "frame". This allows you to view help while scrolling up and down the notebook

```python
?math.log
```
